```python
def reload():
    global df
    x=df.copy()
    return x
```

```python
def run_experiment(df):
    X=df.iloc[:,:-1]
    Y=df.iloc[:,-1]
    Y=pd.DataFrame(Y)
    x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y,test_size=0.2,random_state=100)
    
    #odel = LinearRegression()
    model=RandomForestRegressor(random_state=0)
    model.fit(x_train,y_train)
    
    y_pred = model.predict(x_test)
    a=mean_absolute_error(y_test, y_pred)
    b=metrics.mean_squared_error(y_test, y_pred)
    c=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
    return a,b,c

def run_experiments(preprocesses):
    results = []
    for preprocess in preprocesses:
        df = reload()
        x = preprocess(df)
        if(len(x)==22):
            preprocesses=x
            break
        df=x
        start = time.time()
        a,b,c = run_experiment(df)
        execution_time = time.time() - start
        results.append({
            'name': preprocess.__name__,
            'mean absolute error': a,
            'mean squared error': b,
            'root mean squared error': c,
            'execution time': f'{round(execution_time, 2)}s'
        })
        gc.collect()
    i=1
    for preprocess in preprocesses:
        df=preprocess
        start = time.time()
        a,b,c = run_experiment(df)
        execution_time = time.time() - start
        results.append({
            'name': str(i),
            'mean absolute error': a,
            'mean squared error': b,
            'root mean squared error': c,
            'execution time': f'{round(execution_time, 2)}s'
        })
        gc.collect()
        i=i+1
    df1=pd.DataFrame(results, columns=['name', 'mean absolute error','mean squared error','root mean squared error', 'execution time']).sort_values(by='mean absolute error')
    df2=pd.DataFrame(results, columns=['name', 'mean absolute error','mean squared error','root mean squared error', 'execution time']).sort_values(by='mean squared error')
    df3=pd.DataFrame(results, columns=['name', 'mean absolute error','mean squared error','root mean squared error', 'execution time']).sort_values(by='root mean squared error')
    return df1,df2,df3
```

```python
def original(df):
    return df

def call(df):
    cols=df.columns[:-1]
    l=[]
    for i in cols:
        x=df.copy()
        x.drop(i,axis=1,inplace=True)
        l.append(x)
    return l

def spec(df):
    cols=['Region','Married','Children','cases/1M','Deaths/1M','Age','Infect_Prob']
    df=df.loc[:,cols]
    return df
```

```python
ans1,ans2,ans3=run_experiments([original,spec,call])
```
